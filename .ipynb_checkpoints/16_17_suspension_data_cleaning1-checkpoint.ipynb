```python
# Set up imports
import pandas as pd
import numpy as np
import scipy.stats as stats
import sqlalchemy as sql
import matplotlib.pyplot as plt
```

```python
rawdf = pd.read_csv('res/2016-2017_Student_Discipline_Annual_Report_-_IEP.csv')
```

```python
simplecols = ['dbn', 'locname', 'loccat', 'dist', 'genR', 'genP', 'genS', 'genE', 'swdR', 'swdP', 'swdS', 'swdE', 'total']
rawdf.columns = simplecols
rawdf.head()
```

```python
# normalization --> melt, merge
normdf = rawdf.drop(['locname', 'loccat', 'dist'], axis=1)
normdf = pd.melt(normdf, id_vars=['dbn'])
normdf.rename(columns={'value':'number'}, inplace=True)
        
demo_dict = {'g':'not swd', 't':'all students', 's':'swd'}
normdf['var_name'] = [d[0] for d in normdf['variable']]
normdf['var_name'] = normdf['var_name'].map(demo_dict)

normdf['discipline_type'] = [d[-1] for d in normdf['variable']]

normdf = normdf.drop('variable', axis=1)

print(len(normdf))
print(normdf)
```

```python
rawdf = pd.read_csv('res/2016-2017_Student_Discipline_Annual_Report_-_RACE.csv')
simplecols = [c.lower() for c in rawdf.columns]
rawdf.columns = simplecols

rawdf.drop(['location name', 'location category', 'administrative district'], axis=1, inplace=True)

print(rawdf.loc[1235])
```

```python
racedf = pd.melt(rawdf, id_vars=['dbn'])
racedf['variable'] = racedf['variable'].map(lambda x: x.split(' '))
# racedf['value'] = racedf['value'].map(lambda x: 0 if x == 'R' else x)

racedf['discipline_type'] = [v[-1] for v in racedf['variable']]

discipline_dict = {'removals':'r', 'principal':'p', 'superintendent': 's', 'expulsions':'e', 'removals/suspensions':'all'}
racedf['discipline_type'] = racedf['discipline_type'].map(discipline_dict)

racedf['var_name'] = [' '.join(v[:-1]) for v in racedf['variable']]

racedf['year'] = ['2016' for i in racedf.index]
print(racedf.loc[racedf['dbn'] == '10X045'])
```
